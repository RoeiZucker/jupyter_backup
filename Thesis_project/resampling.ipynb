```python
import pandas as pd
import numpy as np
import os
```

```python
phens = ["Phecode_244.1",
"Phecode_332.0",
"Phecode_426.21",
"Phecode_473.0",
"Phecode_735.2",
"Phecode_737.3",
"Phecode_738.4",
"Phecode_594.3",
"Phecode_594.8",
"Phecode_965.1",
"Phecode_443.9",
"ICD10_M511",
"K409"]
NUMPER_OF_RESAMPLES = 5
DISCRIPTION_FILE = "/cs/usr/roeizucker/my_storage/phewas_run/discription10.csv"
PROJECT_PATH = "/sci/archive/michall/roeizucker/thesis/resampling"
RESULT_PATH = os.path.join(PROJECT_PATH,"phens.csv")
RESULTS_DIR = os.path.join(PROJECT_PATH,"results")
```

```python
pd.set_option('display.max_colwidth', None)
discription_df = pd.read_csv(DISCRIPTION_FILE)
discription_df = discription_df[discription_df["name"].isin(phens)][["name","dataset_path"]]
dfs = []
for index, row in discription_df.iterrows():
    (list(row))
    vals = (list(row))
    
    if vals[0] == "ICD10_M511":
        df =  pd.read_csv( "/cs/labs/michall/roeizucker/phewas_run/runs1/4810:4819/ukbb_dataset.csv")
    else:
        df = pd.read_csv(vals[1])
    num_case_males = df[vals[0] + "_M"].value_counts()[1]
    num_case_females = df[vals[0] + "_F"].value_counts()[1]
    min_case = min(num_case_males,num_case_females)
    phen_to_resample = ""
    if num_case_males < num_case_females:
        phen_to_resample = vals[0] + "_F"
    else:
        phen_to_resample = vals[0] + "_M"
    df1 = df.dropna(subset=[phen_to_resample])
    df1.loc[:,phen_to_resample + "_resampled"] = 0
    df2 = df1[df1[phen_to_resample] == 1]
    df3 = df2.sample(min_case)
    df3[phen_to_resample + "_resampled"] = 1
    df2.loc[:,phen_to_resample + "_resampled"]  = df3[phen_to_resample + "_resampled"]
    df2.loc[:,phen_to_resample + "_resampled"]= df2[phen_to_resample + "_resampled"].fillna(-3)
    df1.loc[:,phen_to_resample + "_resampled"] = df2[phen_to_resample + "_resampled"]
    df1[phen_to_resample + "_resampled"] = df1[phen_to_resample + "_resampled"].fillna(0)
    df1[phen_to_resample + "_resampled"] = df1[phen_to_resample + "_resampled"].replace(-3,None)
    df.loc[:,phen_to_resample + "_resampled"] = df1[phen_to_resample + "_resampled"]
    dfs.append((df,phen_to_resample + "_resampled"))
```
