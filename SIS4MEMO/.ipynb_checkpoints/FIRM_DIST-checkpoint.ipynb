```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
GENE_LIST = ["SH2B3"]
ID_LIST = [1688461,3769417,3813137,4278638]
GENE_EFFECT_FILES_PATH = "/cs/usr/roeizucker/my_storage/virt_env_install_test/ukbb_imputation_gene_effect_scores/"
DATA_PARTICIPANT = "/cs/usr/roeizucker/data_participant.csv"
```

```python
df = pd.read_csv( DATA_PARTICIPANT)
ID_LIST = list(df["eid"])
# REDUCED_ID_LIST = list(df.dropna(subset=["p22402_i2"])["eid"])
REDUCED_ID_LIST = list(df[df["p22402_i2"] > 3]["eid"])
```

```python
gene_table = pd.read_csv("genes_table.csv")
gene_table = gene_table[gene_table["State"] == "Secreted"]

GENE_LIST = list(gene_table["SenMayo human genes"])
# GENE_LIST
```

```python
df = pd.read_csv("/cs/labs/michall/roeizucker/10krun/results/I10_1.csv")[["gene_index","symbol"]]
gene_index_dict = df.set_index("symbol").T.to_dict()
    
```

```python
gene_id = gene_index_dict["IL1A"]['gene_index']
path = os.path.join(GENE_EFFECT_FILES_PATH,str(gene_id) + ".csv")
df = pd.read_csv(path)
over_one = list(filter(lambda x: x>1,list(df["dominant"])))
print(len(df),len(over_one))
```

```python
!head {path}
```

```python
for gene in GENE_LIST:
    if gene not in gene_index_dict.keys():
        continue
    full_path = os.path.join(GENE_EFFECT_FILES_PATH,str(gene_index_dict[gene]['gene_index']) + ".csv")
    df = pd.read_csv(full_path)
    df = df[df["sample_id"].isin(ID_LIST)]
    df_res = df[["dominant","recessive"]]
#     hist = df.hist(bins=30,)
    ax = df_res.plot(kind='hist',subplots=True,sharex=True,sharey=True,title=gene,bins=50)
    plt.title(gene)
    plt.xlim([0,1.5])
    plt.show()
    
    df = df[df["sample_id"].isin(REDUCED_ID_LIST)]
    df_res = df[["dominant","recessive"]]
#     hist = df.hist(bins=30,)
    ax = df_res.plot(kind='hist',subplots=True,sharex=True,sharey=True,title=gene,bins=50)
    plt.title(gene + "_reduced")
    plt.xlim([0,1.5])
    plt.show()
```
