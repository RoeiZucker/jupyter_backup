```python
os.chdir(RESULTS_DIR)
sig_genes = []
for file in glob.glob("E03_F*.csv"):
    df = pd.read_csv(f"{RESULTS_DIR}/{file}")
#     filtered_df = df[df["fdr_significance"]]
    filtered_df = df[df["fdr_qval"] < THREASHOLD]
    sig_genes.append([list(filtered_df["uniprot_id"]),list(filtered_df["symbol"]),list(filtered_df["name"])])
#     print(file)
```

```python
appreances_dict = {}
for genes_list in sig_genes:
    for i,gene in enumerate(genes_list[0]):
        if gene not in appreances_dict:
#             print(len(genes_list))
            appreances_dict[gene] = [0,genes_list[0][i],genes_list[1][i],genes_list[2][i]]
        appreances_dict[gene][0]+=1
```

```python
print(len(appreances_dict))
print(appreances_dict)
male_df = pd.read_csv(f"{RESULTS_DIR}/E03_M_2.csv")
# male_sig_genes = list(male_df[male_df["fdr_significance"]]["uniprot_id"])
male_sig_genes = list(male_df[male_df["fdr_qval"] < THREASHOLD]["uniprot_id"])
print("male_sig_genes:")
for gene in male_sig_genes:
    if gene in appreances_dict:
        print(gene,appreances_dict[gene])
    else:
        print(gene,0)
```

```python
appreances_list = []
for key in appreances_dict:
    appreances_list.append([key,appreances_dict[key]])
appreances_list.sort(key=lambda x:x[1][0],reverse=True)
print(len(appreances_list))
print(appreances_list)
```
