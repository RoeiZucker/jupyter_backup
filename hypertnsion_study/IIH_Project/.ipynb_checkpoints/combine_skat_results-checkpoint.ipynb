# Combine Results
this notebook accepts SKAT pval results per gene and combines it to a more human-friendly result

```python
import pandas as pd
```

```python
SKAT_RESULT_PATH = "/cs/labs/michall/roeizucker/IIH/IIH_results.csv"
BASE_PWAS_DATA = "/cs/labs/michall/roeizucker/phewas_run/results/Phecode_172.2.csv"
BASE_UNIPROT_DATA = "/cs/labs/michall/roeizucker/IIH/uniprot_IIH_SKAT_genes.tsv"
OUTPUT = "/cs/labs/michall/roeizucker/IIH/IIH_results_full.csv"
```

```python
skat_df = pd.read_csv(SKAT_RESULT_PATH)
pwas_df = pd.read_csv(BASE_PWAS_DATA)
uniprot_df = pd.read_csv(BASE_UNIPROT_DATA,sep="\t")
uniprot_df
```

```python
skat_df["SetID2"] = skat_df["SetID"]
pwas_df["SetID2"] = pwas_df["uniprot_id"]
uniprot_df["SetID2"] = uniprot_df["Entry"]

skat_df.set_index("SetID2",inplace=True)
pwas_df.set_index("SetID2",inplace=True)
uniprot_df.set_index("SetID2",inplace=True)
uniprot_df = uniprot_df[~uniprot_df.index.duplicated(keep='first')]
```

```python
skat_df["symbol"] = pwas_df["symbol"]
skat_df["name"] = pwas_df["name"]
skat_df["Involvement in disease"] = uniprot_df["Involvement in disease"]
```

```python
skat_df.to_csv()
```

```python
# create text file with all setID's (uniprot id's)
with open('uniprot_ids.txt',"w") as file:
    for uniprot_id in list(skat_df["SetID"]):
        file.write(uniprot_id + "\n")
!cp uniprot_ids.txt /cs/usr/roeizucker/www/uniprot_ids.txt
```
