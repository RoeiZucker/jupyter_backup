## reads a bim/bed/fam trio

```python
a = pdp.read_plink1_bin(TEST_BED_PATH,TEST_BIM_PATH,TEST_FAM_PATH,verbose=False)

# (bim, fam, bed) = read_plink(join(get_data_folder(), "chr*.bed"),
#                              verbose=False)
```

```python
eid_vals_in_files =  list(a['iid'].values)
```

```python
eid_vals_in_files_set = set(eid_vals_in_files)
```

## reduce a phenotype for non

```python
file = '/cs/labs/michall/roeizucker/10krun/runs/48:59/ukbb_dataset.csv'
target_file = "/cs/labs/michall/roeizucker/SKAT_experiments/I10_data.csv"
target_dir = "/cs/labs/michall/roeizucker/SKAT_experiments"
```

```python
df = pd.read_csv(file)
df = df[df['eid'].isin(eid_vals_in_files)]

df['sort_key'] = df['eid'].astype("str").apply(lambda x: eid_vals_in_files.index(x) if x in eid_vals_in_files_set else float('inf'))

# Sort the DataFrame based on the values in the 'sort_key' column
df = df.sort_values(by='sort_key')


```

```python
df[["eid","I10"]].dropna(subset=['I10']).to_csv(target_file,index=False)
```
