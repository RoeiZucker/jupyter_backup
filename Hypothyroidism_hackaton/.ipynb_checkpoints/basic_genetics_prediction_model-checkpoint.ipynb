```python

cases = []
controls = []
master_cases = None
master_controls = None
for chrom in range(1,22):
    print(chrom)
    chr_df = pd.read_csv(f"/sci/nosnap/michall/roeizucker/Hypothyroidism_hackaton/GWAS/results/e03_reduced_chr{chrom}.PHENO1.glm.logistic",sep="\t")
    requested_variants = list(chr_df[chr_df["P"] < PVAL_RATE]["ID"])
    base_path = PLINK_FILE_BASE_PATH + str(chrom)
    bim_path = base_path + ".bim"
    bed_path = base_path + ".bed"
    fam_path = base_path + ".fam"
    G = pdp.read_plink1_bin(bed_path,bim_path,fam_path,verbose=False)

    for var in requested_variants:
        print(var)
        val_loc = list(G[0]["snp"].values).index(var)
        specific = G.T[val_loc]
        df = specific.to_dataframe()
        df["eid"] = df["iid"]
        df_cases = df[df["eid"].isin(temp_cases_eid)]
        df_controls = df[df["eid"].isin(temp_control_eid)]
        if master_cases is None:
            master_cases = df_cases[["iid"]]
            master_controls = df_controls[["iid"]]
        master_cases.loc[:,var] = df_cases["genotype"]
        master_controls.loc[:,var] = df_controls["genotype"]
master_cases["y"] = 1
master_controls["y"] = 0
master_df = pd.concat([master_cases,master_controls])
```

# run models

```python
def get_gender_X_y(eids,master_cases,master_controls):
    cases = master_cases[master_cases["iid"].astype(str).isin(eids)]
    control = master_controls[master_controls["iid"].astype(str).isin(eids)]
    num = min(len(cases),len(control))
    cases = cases.sample(num)
    control = control.sample(num)
    master_df_partial = pd.concat([cases,control])
    master_df_partial = master_df_partial.dropna(axis=1)
    X = master_df_partial.iloc[:,:-1]
    y = master_df_partial.iloc[:,-1]
    return X,y
```

```python
def perform_multiple_auc_tests_generic(X,y,model,name,params,iter_num):
    catboost_auc = []
    for i in range(iter_num):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.85)
        model = model.fit(X_train.astype(int),y_train.astype(int),**params)
        predictions = model.predict_proba(X_test.astype(int)) 
        auc = roc_auc_score(list(y_test), predictions[:,1])
        catboost_auc.append(auc)
    return catboost_auc

def run_all_models(models_to_run,X,y,iter_num = 90):
    results = []
    for model in models_to_run:
        both_aucs = perform_multiple_auc_tests_generic(X,y,model[0],model[1],model[2],iter_num)
        results.append(both_aucs)
    cols = list(map(lambda x:x[1],models_to_run))
    print(len(cols))
    df = pd.DataFrame(np.array(results).T,columns=cols)
    ax = sns.boxplot(df);
    ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
    plt.show()

def run_all_models_for_all_sex_groups(models_to_run,X_y_pairs,iter_num = 90):
    results = []
    cols = []
    for model in models_to_run:
        for X,y,name in X_y_pairs:
            both_aucs = perform_multiple_auc_tests_generic(X,y,model[0],model[1],model[2],iter_num)
            results.append(both_aucs)
            cols.append(model[1] + "_" + name)
    df = pd.DataFrame(np.array(results).T,columns=cols)
    ax = sns.boxplot(df);
    ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
    plt.show()
    return df

#         print(model[1],"\tmean:", np.array(both_aucs).mean(),"std:", np.array(both_aucs).std())
```

## Calculate AUC for each method

```python
%%time
X = master_df.iloc[:,:-1]
y = master_df.iloc[:,-1]
X_F,y_F = get_gender_X_y(female_eid,master_cases,master_controls)
X_M,y_M = get_gender_X_y(male_eid,master_cases,master_controls)
# shorten names
X_y_pairs = [(X,y,"both"),(X_F,y_F,"female"),(X_M,y_M,"male")]
df = run_all_models_for_all_sex_groups(MODELS_TO_RUN,X_y_pairs)
```
