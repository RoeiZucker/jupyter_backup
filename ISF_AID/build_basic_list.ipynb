```python
import pandas as pd
import numpy as np
```

```python
DISCRIPTION_FILE_PATH = "/cs/labs/michall/roeizucker/phewas_run/discription10.csv"
PHENS =[
    "Phecode_714.1",
    "Phecode_696.4",
    "Phecode_557.1",
    "Phecode_250.1",
    "Phecode_335.0",
    "Phecode_358.1",
    "Phecode_245.0",
    "Phecode_242.1",
    "Phecode_555.1",
    "Phecode_645.22",
    "Phecode_694.2",
    "Phecode_695.42",
    "Phecode_495.0",
    "Phecode_704.1"    
]
ICD10_FILE_PATH = '/cs/labs/michall/roeizucker/phewas_run/phecode_icd10.csv'

RESULT_FILE_NAME = "AID_basic_table_with_name.csv"

for phen in PHENS[:]:
    PHENS.append(phen + "_M")
    PHENS.append(phen + "_F")
```

```python
discription_df = pd.read_csv(DISCRIPTION_FILE_PATH)
discription_df["other_name"] = discription_df["name"].str.replace("_M","")
discription_df["other_name"] = discription_df["other_name"].str.replace("_F","")
discription_df = discription_df.sort_values(by="name")
# discription_df
```

```python
df = discription_df[discription_df["name"].isin(PHENS)]
# print(df)
# print(df)
df = (df.groupby(['other_name','icd_codes','dataset_path'])
      .agg({
          'num_of_cases': lambda x: x.tolist(),
          "num_of_controls":lambda x: x.tolist(),
          "num_relevant_genes":lambda x: x.tolist(),
          
      }))

df[["num_of_cases_B","num_of_cases_F","num_of_cases_M"]] = pd.DataFrame(df["num_of_cases"].tolist(),index=df.index)
df[["num_of_controls_B","num_of_controls_F","num_of_controls_M"]] = pd.DataFrame(df["num_of_controls"].tolist(),index=df.index)
df[["num_relevant_genes_B","num_relevant_genes_F","num_relevant_genes_M"]] = pd.DataFrame(df["num_relevant_genes"].tolist(),index=df.index)

df_final = df.drop(["num_of_cases","num_of_controls","num_relevant_genes"],axis=1).reset_index()
df_final
```

```python
names_df = pd.read_csv(ICD10_FILE_PATH)
names_df = names_df[["PheCode","Phenotype"]]
# names_df["PheCode"] = names_df["PheCode"].astype(str)
names_dict = names_df.set_index("PheCode").to_dict()
print(names_dict["Phenotype"][8.0])
```

```python
df_final["phenotype"] = df_final["other_name"].apply(lambda x: names_dict["Phenotype"][float(str(x).split("_")[1])])
```

```python
df_final.to_csv(RESULT_FILE_NAME)
```

```python
res = []
for val in list(df_final["icd_codes"]):
    res.extend(val.replace("'","").replace("[","").replace("]","").replace(" ","").split(";"))
for i in range(len(res)):
    if len(res[i]) >=4:
        res[i] = (res[i][:3] + "." + res[i][3:])
```

```python
dic = {}
for val in res:
    if val not in dic:
        dic[val] = 0
    dic[val]+=1
# dic
```
