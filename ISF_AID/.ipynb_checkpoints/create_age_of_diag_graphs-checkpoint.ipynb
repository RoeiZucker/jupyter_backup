```python
dates_df = pd.read_csv(DATES_PATH,low_memory=False)
basic_df = pd.read_csv(BASIC_TABLE_PATH)
default_df = pd.read_csv(DEFAULT_DF_PATH)
val = list(default_df.columns[2:12])
default_df = default_df.drop(val,axis=1)
```

```python
cols = []
for index,val in basic_df.iterrows():
#     print(val)
    other_name = val["other_name"]
    print(other_name)
    icd_codes = val["icd_codes"].replace("'","").replace("[","").replace("]","").replace(" ","").split(";")
    dataset_path =  val["dataset_path"]
        
    for i in range(len(icd_codes)):
        if len(icd_codes[i]) >=4:
            icd_codes[i] = (icd_codes[i][:3] + "." + icd_codes[i][3:] + "_date")
        else:
            icd_codes[i]+="_date"

    # get correct eid for phenotype
    phen_df = pd.read_csv(dataset_path)
    eid_list = list(phen_df[phen_df[other_name] == 1]["eid"])
    
    # get minimal date for phen
    end = pd.to_datetime('2099-12-15')
    dates_reduced = dates_df[dates_df["Participant ID"].isin(eid_list)][["Participant ID"] + icd_codes]
    dates_reduced[icd_codes] = dates_reduced[icd_codes].fillna(end)
    for col in icd_codes:
        dates_reduced[col] = pd.to_datetime(dates_reduced[col])
    dates_reduced["min_date_" + other_name] = dates_reduced[icd_codes].min(axis=1)
    dates_reduced.loc[dates_reduced["min_date_" + other_name] == end,"min_date_" + other_name] = None
    
    # marge datasets
    dates_reduced = dates_reduced[["Participant ID","min_date_" + other_name]]
    dates_df = dates_df.merge(dates_reduced, how='left', on='Participant ID')
    cols.append("min_date_" + other_name)
```

```python
dates_df = dates_df[["Participant ID"] + cols]
dates_df["eid"] = dates_df["Participant ID"]
# dates_df.drop("Participant ID",axis=1)
# dates_df
result_df = default_df.merge(dates_df, how='left', on='eid').drop("Participant ID",axis=1)
```

```python
for col in cols:
    result_df["age_"+col] = ((pd.to_datetime(result_df[col]).dt.year -result_df["year_of_birth"])).fillna(0).astype(int)
    result_df["age_"+col].replace(0, np.nan, inplace=True)

```
