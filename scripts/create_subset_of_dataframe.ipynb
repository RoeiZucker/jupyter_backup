```python
import pandas as pd
```

```python
FILE_PATH = "/cs/labs/michall/roeizucker/10krun/runs/48:59/ukbb_dataset.csv"
RESULT_FILE_PATH = "/cs/labs/michall/roeizucker/non_caucasian_run_2/caucasian_reduced_dataset.csv"
PHENOTYPE_TO_KEEP = "I10"
NEW_PHEN_NAME = "I10_NEW"
KEEP_OVERALL = -1
KEEP_SEP = True
KEEP_POS = 591
KEEP_NEG = 5019
```

```python
df = pd.read_csv(FILE_PATH)
```

```python
df[NEW_PHEN_NAME] = None

if KEEP_SEP:
    temp = df[df[PHENOTYPE_TO_KEEP] == 1]
    temp = temp.sample(KEEP_POS)
    temp[NEW_PHEN_NAME] = 1
    df.update(temp)
    
    temp = df[df[PHENOTYPE_TO_KEEP] == 0]
    temp = temp.sample(KEEP_NEG)
    temp[NEW_PHEN_NAME] = 0
    df.update(temp)
else:
    temp = df.sample(KEEP_OVERALL)
    temp.loc[temp[PHENOTYPE_TO_KEEP] == 1,NEW_PHEN_NAME] = 1
    temp.loc[temp[PHENOTYPE_TO_KEEP] == 0,NEW_PHEN_NAME] = 0
    df.update(temp)
df["sample_index"] = df["sample_index"].astype(int)
df["eid"] = df["eid"].astype(int)
df.to_csv(RESULT_FILE_PATH)
```
