```python
ehtnic_contained_searies = df["ehtnic_background_set"].str.contains('|'.join(BACKROUNDS)) & ~(df["ehtnic_background_set"].str.contains(","))

# ehtnic_contained_searies
# df[ehtnic_contained_searies & ~(df["ehtnic_background_set"].str.contains(","))]
```

```python
# TODO: add phen namesx
for phen in TESTED_PHENOTYPES:
    both_sexs_col_name = phen + "_" + ETHNICITY_GROUP_NAME
    male_sexs_col_name = phen + "_M_" + ETHNICITY_GROUP_NAME
    female_sexs_col_name = phen + "_F_" + ETHNICITY_GROUP_NAME
    df[phen + "_" + ETHNICITY_GROUP_NAME] = None
    df[phen + "_M_" + ETHNICITY_GROUP_NAME] = None
    df[phen + "_F_" + ETHNICITY_GROUP_NAME] = None
#     s.str.contains('|'.join(searchfor))
    df.loc[(ehtnic_contained_searies) & (df[phen] == 1.0), f"{phen}_{ETHNICITY_GROUP_NAME}"] = 1.0
    df.loc[(ehtnic_contained_searies) & (df[phen] == 0.0), f"{phen}_{ETHNICITY_GROUP_NAME}"] = 0.0

    df.loc[(ehtnic_contained_searies) & (df[phen+"_M"] == 1.0), f"{phen}_M_{ETHNICITY_GROUP_NAME}"] = 1.0
    df.loc[(ehtnic_contained_searies) & (df[phen+"_M"] == 0.0), f"{phen}_M_{ETHNICITY_GROUP_NAME}"] = 0.0

    df.loc[(ehtnic_contained_searies) & (df[phen+"_F"] == 1.0), f"{phen}_F_{ETHNICITY_GROUP_NAME}"] = 1.0
    df.loc[(ehtnic_contained_searies) & (df[phen+"_F"] == 0.0), f"{phen}_F_{ETHNICITY_GROUP_NAME}"] = 0.0

df.to_csv(os.path.join(PROJECT_PATH,NEW_FILE_NAME))
```
