## add fields with multiple instances

```python
# instances map: 
# n - there are n instances
# 0 - there are no instances
# if the instances start at a specific nunmber, should add a field a - 'starting_index' field
# also need to address array values.
# the printed result should be pasted to the field titels input field in the table exporter.
# the entity type should be participant 
df = pd.read_csv("data_2.csv")
df = df[df["# instances"] != "Singular"]
df = df.dropna(subset=["# instances"])
df['# instances'] = df['# instances'].astype('int')
df = df[df["# instances"] > 1]
df = df.loc[df.index.repeat(df["# instances"])]
df['instance_number'] = df.groupby(['Subject']).cumcount()
df["name"] = df["Subject"] + " | Instance " + df["instance_number"].astype(str)
df.head(10)
fields_list = [x for x in list(df["name"]) if ("," not in x and x not in invalid_names)]
print(len(fields_list))
fields = "Participant ID,"
fields += ','.join(fields_list)
fields += ","
# print("Participant ID,"+ fields)
```

## add fields with a single instances
(this was seperetaed from the previous one, but should be joined)

```python
df = pd.read_csv("data_2.csv")
df = df[df["# instances"] != "Singular"]
df = df.dropna(subset=["# instances"])
df['# instances'] = df['# instances'].astype('int')
df = df[df["# instances"] == 1]
df = df.loc[df.index.repeat(df["# instances"])]
df['instance_number'] = df.groupby(['Subject']).cumcount()
df["name"] = df["Subject"] + " | Instance " + df["instance_number"].astype(str)
df.head(10)
fields_list = [x for x in list(df["name"]) if ("," not in x and x not in invalid_names)]
print(len(fields_list))
fields += ','.join(fields_list)
fields += ","
# print("Participant ID,"+ fields)
```
