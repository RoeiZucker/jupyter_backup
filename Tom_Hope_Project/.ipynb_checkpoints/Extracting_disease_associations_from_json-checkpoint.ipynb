```python
import json
import os
import pandas as pd
import pickle

PATH_TO_DATA = "../../../data/open_targets/associationByOverallDirect/"
PATH_TO_DATA_BY_TYPE_DIRECT = "../../../data/open_targets/by_type/associationByDatatypeDirect/"
PATH_TO_DATA_BY_TYPE_INDIRECT = "../../../data/open_targets/by_type/associationByDatatypeIndirect/"

def extract_json(file_path):
    print(file_path)
    data = [] # a list of dictionaries
    f = open(file_path)
    for line in f:
        data.append(json.loads(line))
#     print(data)
#     print(counter)
    f.close()
    return data
def extract_from_all_files(directory_path):
    all_diseases = []
    for filename in os.listdir(directory_path):
        data = extract_json(directory_path + filename)
        all_diseases += data
    return all_diseases

# unused
def get_disease_ids(associations):
    diseases = []
    for line in associations:
        if(line["diseaseId"] not in diseases):
            diseases.append(line["diseaseId"])
#     print(diseases)
    return diseases

# unused
def get_gene_ids(associations):
    genes = set()
    counter = 0
    for line in associations:
        genes.add(line["targetId"])
        if counter%10000 == 0:
            print(counter)
        counter += 1
#     print(genes)
    return genes
# def get_ids(associations):
#     diseases = []
#     genes = []
#     for i in range(len(associations)):
#         if(associations[i]["diseaseId"] not in diseases):
#             diseases.append(associations[i]["diseaseId"])
#         if(associations[i]["targetId"] not in genes):
#             genes.append(associations[i]["targetId"])
#         print(i)
#     return diseases, genes
def get_ids(associations):
    diseases = set()
    genes = set()
    for i in range(len(associations)):
        diseases.add(associations[i]["diseaseId"])
        genes.add(associations[i]["targetId"])
        if i%10000 == 0:
            print(i)
    return diseases, genes


def filter_genetic_associations(data):
    """returns a list of only dictionaries representing JSON objects with genetic association score."""
    genetic_association_data = []
    for line in data:
        if line["datatypeId"] == "genetic_association":
            genetic_association_data.append(line)
            
    return genetic_association_data
        
    


# diseases = get_disease_ids(all_data)
# print(diseases)

# print(genes) 
# diseases, genes = get_ids(all_data)
def extract_regular_associations():
    all_data = extract_from_all_files(PATH_TO_DATA)
    diseases, genes = get_ids(all_data)
    print("after extraction")
    df = pd.DataFrame(index=list(diseases), columns=list(genes))
    print("after df")
    # print(df)
    counter = 0
    for line in all_data:
    #     df[line["diseaseId"],line["targetId"]] = line["score"]
        df.at[line["diseaseId"],line["targetId"]] = line["score"]
        if counter%1000 == 0:
            print("fill table:" + str(counter))
        counter += 1
    print(df.size)
    print(df)
    print(df.count().mean())
    print(df.count().median())
    df.to_csv("all_disease_associations.csv")
    df.to_pickle("all_disease_associations.pickle")
    print("finished")

def extract_by_type_associations():   
    all_data = extract_from_all_files(PATH_TO_DATA_BY_TYPE_INDIRECT)
    
    filtered_data = filter_genetic_associations(all_data)
    
    diseases, genes = get_ids(all_data)
    print("after extraction")
    df = pd.DataFrame(index=list(diseases), columns=list(genes))
    print(df)
    print("after df")
    counter = 0
    for line in filtered_data:
    #     df[line["diseaseId"],line["targetId"]] = line["score"]
        df.at[line["diseaseId"],line["targetId"]] = line["score"]
        if counter%1000 == 0:
            print("fill table:" + str(counter))
        counter += 1
    print(df.size)
    print(df)
    print(df.count().mean())
    print(df.count().median())
#     df.to_csv("all_disease_genetic_associations.csv")
    df.to_pickle("all_disease_genetic_associations_indirect.pickle")
    print("finished")
    
    
def main():
#     extract_regular_associations()
    extract_by_type_associations()

main()
    
```
