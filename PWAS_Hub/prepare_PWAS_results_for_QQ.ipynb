```python
import pandas as pd
import os
```

```python
FILE_TO_PREPARE = "/cs/labs/michall/roeizucker/10krun/results/I10_1.csv"
GENE_LOCATION_FILE = "/cs/labs/michall/roeizucker/virt_env_install_test/genes_hg19.csv"
OUTPUT_FILE = "./I10_1_with_locations.csv"
```

```python
df_genes = pd.read_csv(GENE_LOCATION_FILE)
df_data = pd.read_csv(FILE_TO_PREPARE)
```

```python
genes_dict = df_genes[["uniprot_id","cds_start","chr"]].set_index("uniprot_id").to_dict()
cds_start_dict = genes_dict["cds_start"]
chr_dict = genes_dict["chr"]
```

```python
def create_file(chr_dict,cds_start_dict,file_to_prepare,output_path):
    df_data = pd.read_csv(file_to_prepare)
    df_data["POS"] = df_data["uniprot_id"].map(cds_start_dict)
    df_data["#CHROM"] = df_data["uniprot_id"].map(chr_dict)
    df_data["P"] = df_data["combined_pval"]
    df_data.to_csv(output_path)
def list_specific_files_in_directory(directory_path):
    file_list = []
    # Walk through the directory
    for root, dirs, files in os.walk(directory_path):
        for file in files:
            # Check if the file ends with '_1.csv'
            if file.endswith('_1.csv'):
                # Add each matching file's full path to the list
                file_list.append(os.path.join(root, file))
    return file_list

# Example usage
directory_path = '/cs/labs/michall/roeizucker/10krun/results/'
files = list_specific_files_in_directory(directory_path)
for file in files:
    print(file)
    output_path = file.replace(".csv","_with_locations.csv")
    create_file(chr_dict,cds_start_dict,file,output_path)
```
