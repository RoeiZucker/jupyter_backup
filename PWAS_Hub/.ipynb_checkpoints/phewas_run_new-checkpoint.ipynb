## Prepare phenotypes

```python
def set_to_dict(dic,key,val,source):
    dic[key] = [val,source]
    pass
file1 = pd.read_csv(ICD10_FILE_PATH)
phewas_dict = {}

for index in file1.index:
    curr_id = file1["PheCode"][index]
    curr_name = file1["Phenotype"][index]
    if str(file1["PheCode"][index]) == "nan":
#         set_to_dict(icd10_dict,file1["ICD10"][index],file1["ICD10 String"][index],"Phewas_icd10")
        phewas_dict[file1["ICD10"][index]] = [file1["ICD10 String"][index],"Phewas_icd10",[file1["ICD10"][index]]]
    else:
        if file1["PheCode"][index] not in phewas_dict:
            phewas_dict[file1["PheCode"][index]] = [file1["Phenotype"][index],"Phewas_phecode",[]]
        if "." in file1["ICD10"][index]:
            phewas_dict[file1["PheCode"][index]][-1].append(file1["ICD10"][index])
        
        
```
