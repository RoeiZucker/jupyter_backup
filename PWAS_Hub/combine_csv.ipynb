```python
import pandas as pd
```

```python
PATH = "/cs/labs/michall/roeizucker/phewas_run/description12.csv"
```

```python
df = pd.read_csv(PATH)
df["name"] = df["name"].str.replace("_M$","",regex=True).str.replace("_F$","",regex=True)
df
```

```python
# returns Both/Male/Female
def split_col(row,col="num_relevant_genes"):
    
#     print(row)
    
    if row is None or row["sex"] is None or "B" not in row["sex"] or "M" not in row["sex"] or "F" not in row["sex"]:
        return ["-1","-1","-1"]
    pos_B = int(row["sex"].index("B"))
    pos_M = int(row["sex"].index("M"))
    pos_F = int(row["sex"].index("F"))
#     print([row[col][pos_B],row[col][pos_M],row[col][pos_F]])
#     print(row)
    return [row[col][pos_B],row[col][pos_M],row[col][pos_F]]

cols = ['result_file_path',"dataset_path",'sex',"num_of_cases","num_of_controls","num_relevant_genes","num_relevant_genes_0.1","MF_prevalence_ratio","FM_prevalence_ratio"]
df_reversed = df.groupby(['name',"meaning"])[cols].agg(list).reset_index()
for col in cols:
    if col == "sex":
        continue
    print(col)
#     df_reversed[[col + "_B", col + "_M", col +"_F"]] = df_reversed.apply(lambda x:split_col(x,col) ,axis=1)
    df_reversed[col] = df_reversed.apply(lambda x:split_col(x,col) ,axis=1)
    df_reversed[[col + "_B", col + "_M", col +"_F"]] = pd.DataFrame(df_reversed[col].tolist(), index= df_reversed.index)

#     df_reversed[[col + "_B", col + "_M", col +"_F"]] = df_reversed[col].str.split(' ', expand=True)
```

```python
df_reversed.to_csv("description_12_combined.csv")
```
