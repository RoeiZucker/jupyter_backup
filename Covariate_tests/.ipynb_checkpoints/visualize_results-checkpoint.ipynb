```python
import matplotlib.pyplot as plt

# We get the unique covar_types in sorted order
unique_covars = sorted(df["covar_type"].unique())

# For each covar_type, collect the found_overall values as a list
data_for_boxplot = [
    df.loc[df["covar_type"] == cov, DESIERED_PARAMETER].dropna().values
    for cov in unique_covars
]

plt.figure(figsize=(8, 6))
plt.boxplot(data_for_boxplot, labels=unique_covars, showmeans=True)
plt.xticks(rotation=45, ha="right")
plt.ylabel(DESIERED_PARAMETER_NAME)
plt.title(f"Distribution of {DESIERED_PARAMETER_NAME} by Covariate Type for {VERSION_NAME}")
plt.tight_layout()
plt.savefig(f"{DESIERED_PARAMETER}_{VERSION}.png")
plt.show()
```

```python
results = []
for her_score in HERETABILITY_SCORES:
    for cov_score in COVARIATE_SCORES:
        if her_score + cov_score > 1:
            continue
        for covar_type in COVAR_GROUPS:
            try:
                phenotype_file_result_name = (
                    f"generated_pehn_result_ver={VERSION}_heretability={her_score}_covar_effect={cov_score}")
                phenotype_file_result_name = phenotype_file_result_name+".txt"
                
                result_path = os.path.join(PLINK_RESULT_DIR,phenotype_file_result_name.replace(".txt",covar_type) + f".log")
                pattern = re.compile("\(based on median chisq\) =(.*)")
                val = -100.0
                for i, line in enumerate(open(result_path)):
                    for match in re.finditer(pattern, line):
                        a = match.group()
                        val = float(a.split()[-1][:-1])
                        break

                results.append(
                    [her_score,cov_score,covar_type,val]
                )
                # print(variants)
            except OSError as exc:
                print(exc)
                print(covar_file,"failed")
            except Exception as exc:
                print(exc)
                print(covar_type,"check if done")
```

```python
cols = ["her_score","cov_score","covar_type",
        "lambda"]
res_df = pd.DataFrame(results,columns=cols)
res_df
```

```python
import matplotlib.pyplot as plt

# We get the unique covar_types in sorted order
unique_covars = sorted(res_df["covar_type"].unique())

# For each covar_type, collect the found_overall values as a list
data_for_boxplot = [
    res_df.loc[res_df["covar_type"] == cov, "lambda"].dropna().values
    for cov in unique_covars
]

plt.figure(figsize=(8, 6))
plt.boxplot(data_for_boxplot, labels=unique_covars, showmeans=True)
plt.xticks(rotation=45, ha="right")
plt.ylabel(DESIERED_PARAMETER)
plt.title(f"Distribution of lambda by Covariate Type for {VERSION_NAME}")
plt.tight_layout()
plt.savefig(f"lambda_{VERSION}.png")

plt.show()
```
