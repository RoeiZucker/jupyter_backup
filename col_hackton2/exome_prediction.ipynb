```python
import pandas as pd
import os
from IPython.display import clear_output
```

```python
BATCH_SIZE = 100000
```

```python
# 3381

df = pd.read_csv("/sci/archive/michall/roeizucker/col_hackaton/batch_sample_LDL_ids.aa_ad_547.g.vcf.gz",compression='gzip',skiprows=3381,nrows=100000,sep="\t")
# df = pd.read_csv("/sci/archive/michall/roeizucker/col_hackaton/batch_sample_LDL_ids.aa_ad_1..547.g.vcf",skiprows=3381,sep="\t")
```

```python
threshold = 0.10
count = 0
# pd.read_csv('test.csv', sep='|', skiprows=range(1, 10))
# skiprows=count*BATCH_SIZE + 3381
while len(df) > 10:
    clear_output(wait=True)
    print(count)
    df = pd.read_csv("/sci/archive/michall/roeizucker/col_hackaton/batch_sample_LDL_ids.aa_ad_547.g.vcf.gz",compression='gzip',skiprows=list(range(3381)) + list(range(3382,count*BATCH_SIZE + 3382)),nrows=BATCH_SIZE,sep="\t")

# Count the occurrences of "." in each row
    dot_counts = df.eq('.').sum(axis=1)

    # Identify rows to remove (more than threshold percentage of "." values)
    rows_to_remove = dot_counts > (df.shape[1] * threshold)

    # Filter the dataframe to keep only desired rows
    filtered_df = df[~rows_to_remove]
    filtered_df.to_csv(f"/sci/archive/michall/roeizucker/col_hackaton/batch_sample_LDL_ids{count}.csv")
    count += 1

```

```python
filtered_dfs = []
i = 0
file_name = f"/sci/archive/michall/roeizucker/col_hackaton/batch_sample_LDL_ids{i}.csv"
while os.path.exists(file_name):
    if i % 10 == 0:
        clear_output(wait=True)
        print(i)
    filtered_dfs.append(pd.read_csv(file_name))
    i+=1
    file_name = f"/sci/archive/michall/roeizucker/col_hackaton/batch_sample_LDL_ids{i}.csv"
```

```python
master_df = pd.concat(filtered_dfs)
```
