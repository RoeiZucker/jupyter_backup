```python
def find_relevant_interval(chrom,loc):
    return list(interval_df[(interval_df[0] == chrom) & (interval_df[1] < loc + 1) & (interval_df[2] > loc - 1)][3])
# interval_df
# print(find_relevant_interval("chr1",978629))
```

```python
featuers = pd.DataFrame(users).T
UTR_cols = [col for col in featuers.columns if 'UTR' in col]
featuers = featuers[UTR_cols]
featuers
```

```python
eid_translator = pd.read_csv(EID_TRANSLAT0R,sep=" ",header=None)
eid_translator.set_index(1,inplace=True)
eid_dict = eid_translator.to_dict()[0]
```

```python
pred_df = pd.read_csv(PRED)
```

```python
pred_df
```

```python
y = []

for row in featuers.iterrows():
    eid = (eid_dict[row[0]])
    val = pred_df[pred_df["Participant ID"] == eid]
    y.append(val["Average Diameter for LDL Particles | Instance 0"].iloc[0])
```

```python
from sklearn.linear_model import LinearRegression
y1 = np.array(y)
y1[np.isnan(y1)] = 0
reg = LinearRegression().fit(featuers.fillna(0), y1)
```
